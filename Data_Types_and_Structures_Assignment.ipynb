### 1. What are data structures, and why are they important?

Data structures are ways to store and organize data so we can use it easily. They are important because they help us write better and faster programs.

### 2. Explain the difference between mutable and immutable data types with examples

Mutable data types can change their content without changing their identity. Example: list.
```python
my_list = [1, 2, 3]
my_list.append(4)
```
Immutable types cannot be changed after creation. Example: string.
```python
my_str = 'hello'
# my_str[0] = 'H'  # This gives an error
```

### 3. What are the main differences between lists and tuples in Python?

Lists are mutable, meaning we can change them. Tuples are immutable, meaning we cannot change them.

### 4. Describe how dictionaries store data

Dictionaries store data in key-value pairs. Each key is unique and is used to get its value.

### 5. Why might you use a set instead of a list in Python?

Sets do not allow duplicate values, so they are good when we want only unique items.

### 6. What is a string in Python, and how is it different from a list?

A string is a sequence of characters. It is immutable. A list can hold any data types and is mutable.

### 7. How do tuples ensure data integrity in Python?

Since tuples cannot be changed, they help keep the data safe from accidental changes.

### 8. What is a hash table, and how does it relate to dictionaries in Python?

A hash table stores data using a key. Dictionaries in Python use hash tables internally.

### 9. Can lists contain different data types in Python?

Yes, lists can have different types like integers, strings, and even other lists.

### 10. Explain why strings are immutable in Python

Strings are immutable to make them safer and to help with performance.

### 11. What advantages do dictionaries offer over lists for certain tasks?

Dictionaries allow faster lookup using keys, while lists require searching by index.

### 12. Describe a scenario where using a tuple would be preferable over a list

When the data should not change, like coordinates (x, y), we should use a tuple.

### 13. How do sets handle duplicate values in Python?

Sets automatically remove duplicate values. Only unique items are kept.

### 14. How does the “in” keyword work differently for lists and dictionaries?

For lists, 'in' checks for a value. For dictionaries, 'in' checks for a key.

### 15. Can you modify the elements of a tuple? Explain why or why not

No, because tuples are immutable. Their values cannot be changed.

### 16. What is a nested dictionary, and give an example of its use case

A nested dictionary is a dictionary inside another dictionary.
```python
data = {'person': {'name': 'John', 'age': 30}}
```

### 17. Describe the time complexity of accessing elements in a dictionary

It is O(1), which means it is very fast and constant time.

### 18. In what situations are lists preferred over dictionaries?

When we only care about values and their order, not keys.

### 19. Why are dictionaries considered unordered, and how does that affect data retrieval?

Earlier Python versions had unordered dictionaries. Now they keep order, but we still access values by keys.

### 20. Explain the difference between a list and a dictionary in terms of data retrieval.

In a list, we use indexes to get data. In a dictionary, we use keys.
