## Data Types and Structures - Theory Questions (As a New Learner)

### 1. What are data structures, and why are they important?

Data structures are ways to organize and store data. They are important because they help us manage and use data efficiently. For example, lists and dictionaries in Python help us keep data in a way that is easy to access and change.

### 2. Explain the difference between mutable and immutable data types with examples

Mutable types can be changed. For example, lists can have new items added.
```python
my_list = [1, 2, 3]
my_list.append(4)  # Now it's [1, 2, 3, 4]
```
Immutable types can't be changed. For example, strings and tuples can't be changed after creation.
```python
my_str = 'hello'
# my_str[0] = 'H'  # This gives an error because strings are immutable
```
